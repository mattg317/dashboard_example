```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in cleaned dataset
cleaned_df = pd.read_csv("../CleanData/complete_data.csv", index_col=0)
cleaned_df.head()
```

```python
# Group by class
classes_df = cleaned_df.groupby("SectionName") \
                        .agg({"Mastery": "mean",
                              "Support": "mean",
                             'Pace': 'mean',
                             'InstructorKnowledge': 'mean',
                             'InstructorClarity': 'mean',
                             'InstructorEngagement': 'mean'})
classes_df.head()
```

```python
# Group by university
university_df = cleaned_df.groupby("UniversityName") \
                        .agg({"Mastery": "mean",
                              "Support": "mean",
                             'Pace': 'mean',
                             'InstructorKnowledge': 'mean',
                             'InstructorClarity': 'mean',
                             'InstructorEngagement': 'mean'})
university_df.head()
```

### Class Heatmaps
* Split into groups for readability

```python
first_group = classes_df.iloc[:25]
sns.heatmap(first_group, annot=True, cmap="RdYlGn")
```

```python
second_group = classes_df.iloc[25:50]
sns.heatmap(second_group, annot=True, cmap="RdYlGn")
```

```python
last_group = classes_df.iloc[50:75]
sns.heatmap(last_group, annot=True, cmap="RdYlGn")
```

### University Heatmap

```python
sns.heatmap(university_df, annot=True, cmap="RdYlGn")
```
