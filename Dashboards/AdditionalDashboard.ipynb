```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
cleaned_df = pd.read_csv("../CleanData/complete_data.csv", index_col=0)
cleaned_df.head()
```

```python
# Group by class
classes_df = cleaned_df.groupby("SectionName") \
                        .agg({"Mastery": "mean",
                              "Support": "mean",
                             'Pace': 'mean',
                             'InstructorKnowledge': 'mean',
                             'InstructorClarity': 'mean',
                             'InstructorEngagement': 'mean'})
classes_df.head()
```

```python
# Group by university
university_df = cleaned_df.groupby("UniversityName") \
                        .agg({"Mastery": "mean",
                              "Support": "mean",
                             'Pace': 'mean',
                             'InstructorKnowledge': 'mean',
                             'InstructorClarity': 'mean',
                             'InstructorEngagement': 'mean'})
university_df.head()
```
