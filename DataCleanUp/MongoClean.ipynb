```python
import pandas as pd
import numpy as np
```

```python
# Files
classes = "../Resources/classesCollection.xlsx"
feedback = "../Resources/studentFeedbackFormsCollection.xlsx"
university = "../Resources/universitiesCollection.xlsx"
```

```python
# Store class data into DataFrame
classes_df = pd.read_excel(classes)
classes_df.head()
```

```python
# Store feedback data in DataFrame
feedback_df = pd.read_excel(feedback)
feedback_df.head()
```

```python
# Select needed columns
feedback_df.columns = ['_id', 'week', 'Pace', 'Support', 'Mastery', 'InstructorEngagement', \
                    'InstructorClarity', 'InstructorKnowledge', 'submitted', 'universityId', 'class_id']

# Drop data with no responses
feedback_df = feedback_df.dropna(thresh=6)
feedback_df.head()
```

```python
len(feedback_df)
```

```python
# Store University data into DataFrame
university_df = pd.read_excel(university)
university_df.head()
```

```python
# Merge feedback and class DataFrames
class_student_df = pd.merge(feedback_df, classes_df, left_on='class_id', right_on="_id")
class_student_df.head()
```

```python
len(class_student_df)
```

```python
# Merge university DataFrame into previous DataFrame
full_df = pd.merge(class_student_df, university_df, left_on='universityId_x', right_on='_id')
full_df.head()
```

```python
len(full_df)
```

```python
# Check Columns
full_df.columns
```

```python
# Select columns
full_clean_df = full_df[['code', 'Pace', 'Support', 'Mastery', 'InstructorEngagement', 'InstructorClarity', 
                         'InstructorKnowledge', 'name']]
full_clean_df.head()
```

```python
# Rename columns
full_clean_df.columns = ["SectionName", "Pace", "Support", "Mastery", 'InstructorEngagement', 'InstructorClarity', 
                         'InstructorKnowledge', 'UniversityName']
full_clean_df.head(15)
```

```python
# Reorder columns
reorder_df = full_clean_df[['SectionName', 'UniversityName', "Pace", "Support", "Mastery", 'InstructorEngagement', 'InstructorClarity', 
                         'InstructorKnowledge']]
reorder_df.head()
```

```python
len(reorder_df)
```

### Handling missing data

```python
# See how much data is missing
missing_values_count = reorder_df.isnull().sum()
missing_values_count
```

```python
# Determine total missing values
total_cells = np.product(reorder_df.shape)
total_missing = missing_values_count.sum()

# Percent of data that is missing
percent_missing = (total_missing/total_cells) * 100

print(f"Total missing: {total_missing}")
print(f"Percent missing: {percent_missing}")
```

```python
# Drop rows with Nan Values
mongo_complete_df = full_clean_df.dropna()
mongo_complete_df = mongo_complete_df[['SectionName', 'UniversityName', "Pace", "Support", "Mastery", 'InstructorEngagement', 'InstructorClarity', 
                         'InstructorKnowledge']]
mongo_complete_df.head()
```
