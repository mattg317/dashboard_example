```python
import pandas as pd
```

```python
# Files
classes_table = "../Resources/classes_table.csv"
student_feedback_table = "../Resources/student_feedback_table.csv"
university_table = "../Resources/universities_table.csv"
```

```python
# Store classes csv into DataFrame
classes_df = pd.read_csv(classes_table)
classes_df.head()
```

```python
# Store studentfeedback csv into DataFrame
student_feedback_df = pd.read_csv(student_feedback_table)
student_feedback_df.head()
```

```python
print(len(student_feedback_df))
```

```python
# Store University csv into DataFrame
university_df = pd.read_csv(university_table)
university_df.head()
```

```python
# Merge classes DataFrame with Student Feedback DataFrame
class_student_df = pd.merge(classes_df, student_feedback_df, on='SectionID')
class_student_df.head()
```

```python
print(len(class_student_df))
```

```python
# Merge results of last DataFrame with the University DataFrame
class_student_uni_df = pd.merge(class_student_df, university_df, on="UniversityID", how="left")
class_student_uni_df.head()
```

```python
print(len(class_student_uni_df))
```

```python
# View columns to see what to pull out
class_student_uni_df.columns
```

```python
# Filter columns needed
clean_df = class_student_uni_df[['SectionName','UniversityName', 'InstructorEngagement', \
                     'PaceOfClassThisWeek', 'AcademicSupport', 'MasteredConceptsCovered', \
                     'InstructionsClarity', 'InstructorKnowledge']]
clean_df.head()
```

```python
# Rename columns for clarity
clean_complete_df = clean_df.rename(columns={"PaceOfClassThisWeek": "Pace",
                                            "AcademicSupport": "Support",
                                            "MasteredConceptsCovered": "Mastery",
                                            "InstructionsClarity": "InstructorClarity"})
clean_complete_df.head()
```

```python
# Reorder columns
sql_complete_df = clean_complete_df[['SectionName', 'UniversityName', "Pace", "Support", "Mastery", 'InstructorEngagement', 'InstructorClarity', 
                         'InstructorKnowledge']]
sql_complete_df.head()
```

```python
len(sql_complete_df)
```

```python
# Export CSV for final truth of data
sql_complete_df.to_csv("../CleanData/sql_feedback.csv")
```
